```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

matplotlib.rc('font', family='BIZ UDGothic')
```

```python
data = np.load('06-2024npCPI.npz', allow_pickle = True)
print(data.files)
```

```python
index = data['index']
x = data['months']
cpi = data['data']
```

```python
print(index, '\n', len(index))
print(cpi.shape)
```

```python
fig = plt.figure(figsize = (10, 6)) 
ax = fig.add_subplot(1,1,1)

l = [0, 6, 9, 5]
for i in l:
    ax.plot(x, cpi[:, i], linewidth = .8, label = index[i + 1])
ax.set_ylim(0, 160)
ax.set_xlabel("Year")
ax.set_ylabel("消費者物価指数")
ax.legend(loc = 'lower right')
ax.spines['right'].set_visible(False) 
ax.spines['top'].set_visible(False) 
plt.savefig("CPI-long.png")   
plt.show()
```

```python
start = 560

fig = plt.figure(figsize = (8, 6)) 
ax = fig.add_subplot(1,1,1)

l = [0, 6, 9, 5]
for i in l:
    ax.plot(x[start:], cpi[start:, i], linewidth = .8, label = index[i + 1])
ax.set_ylim(80, 140)
ax.set_xlabel("Year")
ax.set_ylabel("消費者物価指数")
ax.legend(loc = 'lower right')
ax.spines['right'].set_visible(False) 
ax.spines['top'].set_visible(False) 
plt.savefig("CPI-short.png")   
plt.show()
```
